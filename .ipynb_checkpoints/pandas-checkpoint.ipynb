```python
newdf1
```

```python
#index 0 row has deleted (axis = 0 is row, axis = 1 is column)
newd = newdf1.drop(0, axis=0)
```

```python
#Column 'A' will be deleted 
newdf = newdf1.drop('A', axis=1)
```

```python
# row number 3 to 4 and column A and B
newdf.loc[[3,4],['B','C']]
```

```python
# all rows and b , and c column
newdf.loc[:,['B','C']]

```

```python
# all rows and columns
newdf.loc[:,:]
```

```python
# b values grater than 0.3 adn c > 0.5
newdf.loc[(newdf['B']>0.3) & (newdf['C']>0.5)]
```

```python
# iloc is used to get data only based on counting, whereas loc is used to pull data with indax name
newdf.iloc[1,3]
```

```python
newdf.iloc[[0,4],[0,3]]
```

```python
# row 0 deleted
newdf.drop(0)
```

```python
# return copy, if you want to change in existing, use inplace=true
newdf.drop(['B','C'], axis=1, inplace=True)
```

```python
# return copy, if you want to change in existing, use inplace=true
newdf.drop([0,2,4], axis=0, inplace=True)
```

```python
# reset index but this will also add extra index column, if we dont want this,  drop=True
newdf.reset_index(drop=True)
```

```python
# get null info
newdf.isnull()
newdf['B'].isnull()
```

```python
# fill None in any column, its not good way, use loc function to change value
#newdf['B'] = None (not right way)
newdf.loc[:, ['B']] = 56 # (right way)
```

```python
newdf
```

```python
# remove NaN data
newdf.dropna(how='all', axis=1)
```

```python
# delete duplicate, it doesn't support axis, instead it work with subset
newdf.drop_duplicates(subset='B')
```
